```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
dframe = DataFrame({'key':['X','Z','Y','Z','X','X'], 'data_set_1':np.arange(6)})
dframe
```

```python
dframe1 = DataFrame({'key':['Q','Y','Z'], 'data_set_1':[1,2,3]})
dframe1
```

```python
pd.merge(dframe, dframe1)
```

```python
pd.merge(dframe,dframe1, on='key')
```

```python
pd.merge(dframe,dframe1, on='key', how='left')
```

```python
df_left = DataFrame({'key1':['SF','SF','LA'], 'key2':['one','two','one'], 'left_data':[10,20,30]})
```

```python
df_right = DataFrame({'key1':['SF','SF','LA','LA'], 'key2':['one','one','one','two'], 'right_data':[40,50,60,70]})
```

```python
df_left
```

```python
df_right
```

```python
pd.merge(df_left,df_right,on=['key1','key2'], how='outer')
```

```python
pd.merge(df_left,df_right,on='key1')
```
